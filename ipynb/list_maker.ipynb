```python
def feature_list_maker(dataframe):
    df_test = dataframe.drop(columns=['msrp','vin','color',])
    #names for each DF/ Model to be trained


    # divide the data
    X = df_test.drop(columns=['ask_price'])
    y = df_test['ask_price']

    # set dummies
    X_dummies = pd.get_dummies(X,drop_first=True)

    # split for test and train
    feature_list = X_dummies.columns.to_list()
    return feature_list
```

```python
sedan_feature_list = feature_list_maker(sedan_df)
suv_feature_list = feature_list_maker(suv_df)
trucks_feature_list = feature_list_maker(trucks_df)
full_feature_list = feature_list_maker(full_df)
```
